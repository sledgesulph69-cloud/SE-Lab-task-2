```python
from typing import List

class Solution:
    def countGreaterThanAverage(self, nums: List[int]) -> int:
        if not nums:
            return 0

        avg = sum(nums) / len(nums)
        count = 0
        for x in nums:
            if x > avg:
                count += 1
        return count
```

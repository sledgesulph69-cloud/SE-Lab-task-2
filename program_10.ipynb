```python
def pascal_triangle(n):
    pas_triangle = [[1]]
    for i in range(1, n+1):
        row = [0]*(i+1)
        row[0] = row[-1] = 1
        pas_triangle.append(row)
        for j in range(1, i):
            pas_triangle[i][j] = pas_triangle[i-1][j-1]+pas_triangle[i-1][j]
    return pas_triangle
ncr = pascal_triangle(1000)
y = 1
while y:
    n = int(input("enter value of n: "))
    r = int(input("enter value of r" ))
    print(ncr[n][r])
    y = int(input("do u want to continue: "))
```

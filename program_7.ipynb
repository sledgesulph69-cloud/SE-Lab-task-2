```python
#write a program to find factorial in a range 1 to 100 the result should be modulo arithmetic wer prime number p is 10**9+7
def mult(a,b,p):
    return (a%p * b%p)%p

n = 100
fact = 1
p = 10**9+7
for i in range(1, n+1):
    fact = mult(fact, i, p)

print(fact)
```

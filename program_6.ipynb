```python
def div(x, y, p):
    return (x%p)*(power(y,p-2, p)) % p
def power(x,y,p):
    if y == 1:
        return x
    res = power(x,y//2, p)
    res = (res%p)*(res%p)
    if y%2 == 1:
        res = (res%p)*(x%p)
    return res
print(div(100,20, 3))
```

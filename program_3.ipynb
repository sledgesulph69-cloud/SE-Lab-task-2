```python
dp = {
    0:1,
    1:1
}
def catalan_w_dp(n):
    if n not in dp:
        dp[n] = 0
        for i in range(n):
            dp[n] += catalan_w_dp(i)*catalan_w_dp(n-i-1)
    return dp[n]
for n in range(11):
    print(n, catalan_w_dp(n))
```

```python
# write a program to find the number of tosses requires to get n consecutive heads

import random

def num_tosses_using_formula(n):
    return 2**(n+1) - 2

def num_tosses_using_simulation(n):
    num_of_trails = 10000
    total_no_of_tosses = 0
    for i in range(num_of_trails):
        consecutive_heads = 0
        num_of_tosses = 0
        while(consecutive_heads < n):
            toss = random.randint(1, 7)
            num_of_tosses += 1
            if toss == 1:
                consecutive_heads += 1
            else:
                consecutive_heads = 0
        total_no_of_tosses += num_of_tosses
    return total_no_of_tosses/num_of_trails

n_values = list(range(2,11))
for n in n_values:
    print(n, num_tosses_using_formula(n),num_tosses_using_simulation(n))
```

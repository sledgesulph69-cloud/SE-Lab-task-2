```python
def mult(a, b, p):
    return (a % p * b % p) % p

def expm(x, y, p):
    if y == 0:
        return 1
    if y == 1:
        return x % p
    half = expm(x, y // 2, p)
    res = mult(half, half, p)
    if y % 2 == 1:
        res = mult(res, x, p)
    return res

a = 3
b = [1, 2, 3, 4]
p = 1337
b = b[::-1]  # reverse the list
res = 1

for digit in b:
    # Compute: res = mult(res, expm(res, expm(a, digit, p)), p)
    inner_exp = expm(a, digit, p)        # a^digit mod p
    outer_exp = expm(res, inner_exp, p)  # res^(a^digit) mod p
    res = mult(res, outer_exp, p)
    a = expm(a, 10, p)  # a = a^10 mod p

print(res)
```

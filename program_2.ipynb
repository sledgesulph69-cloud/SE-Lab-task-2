```python
#find the no of persons required to be present where the probabilty p to have two persons with the same bday
persons = 1
pcomp = 1
p = 0.5
numerator = 364
while ((1-pcomp)<p):
    pcomp*=numerator/365
    numerator -=1
    persons += 1
    print(1-pcomp, persons)
```
